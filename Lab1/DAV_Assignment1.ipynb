# 1. Create two data frames by reading given two files.

Data Frame using "district_level_manufacturing_msme.json" file

```python
filepath = "/content/drive/MyDrive/data/district_level_manufacturing_msme.json"
f = open(filepath)
data = json.load(f)

columns = pd.DataFrame(data["fields"])["label"].tolist()
manufacture_df = pd.DataFrame(data["data"], columns = columns)

print(manufacture_df.shape) #(row, column) tuple indicating  rows and columns in the data
print()
manufacture_df.info()  #returns the information about the columns
```

```python
manufacture_df.head()  #prints first 5 rows
```

```python
#since several columns have integer value but from the column info
#we can see that it of type object
#hence, we will convert it into integers

cols = ['Lg_Dist_Code', 'MICRO', 'SMALL', 'MEDIUM', 'Total']
manufacture_df[cols] = manufacture_df[cols].astype(int)
```

Data Frame using "district_level_service_msme.csv" file

```python
filepath = "/content/drive/MyDrive/data/district_level_service_msme.csv"
service_df = pd.read_csv(filepath)

print(service_df.shape) 
print()
service_df.info()
```

```python
service_df.head() 
#we don't need to change the data type of columns as it is already integer
```

# 2. Find out total "Small" Manufacturing MSME in India.

```python
manufacture_df['SMALL'].sum()
```

# 3. Create a dataframe having state wise total number of "Micro", "Small" and "Medium" Services MSE and save the results as a CSV file.

```python
df = service_df.groupby('STATE_NAME').sum()[['MICRO', 'SMALL', 'MEDIUM']]
df.head()
```

```python
#saving the new dataframe as csv
df.to_csv("/content/drive/MyDrive/data/state_wise_service_msme.csv")
```

# 4. Join the both the data frame based on common STATE NAME, DISTRICT NAME, Lg Dist Code and Last Updated. The result should look like below. "x" and "y" represent the manufacturing MSME and service MSME respectively.

```python
cols = ['STATE_NAME', 'DISTRICT_NAME', 'Lg_Dist_Code', 'Last_Updated']
df1 = pd.merge(left=manufacture_df, right=service_df, on=cols)
df1.head()
```

# 5. Create a Pivot Table having rows STATE NAME and columns Service and Manufacturing "MSME" as show in below. Use "Sum" to add up district wise number.

```python
pivot_msme = pd.pivot_table(df1, index = 'STATE_NAME', aggfunc = np.sum)
pivot_msme = pivot_msme.drop(['Lg_Dist_Code', 'Total_x', 'Total_y'], axis=1)
pivot_msme.head()
```
